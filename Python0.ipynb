- Sabrati, podeliti, oduzeti i pomnožiti promenljive, i proveriti rezultujući tip

```python
Mile = 10
Žika = 4.2

zbir = Mile + Žika
razlika = Mile - Žika
proizvod = Mile * Žika
kolicnik = Mile / Žika

print("Zbir:", zbir, type(zbir))
print("Razlika:", razlika, type(razlika))
print("Proizvod:", proizvod, type(proizvod))
print("Količnik:", kolicnik, type(kolicnik))
```

- Sabrati njegov prvi i poslednji element i smestiti ga u promenljivu *b*
- Pomnožiti sve elemente niza *a* promenljivom *b*
- Proveriti tip promenljive *b*
- Ponovo proveriti tip niza *a*

```python
a = [9, 8, 4, 7, 1]

# sabiranje prvog i poslednjeg elementa
b = a[0] + a[-1]

# množenje svih elemenata niza a promenljivom b
a = [x * b for x in a]

# provera tipa promenljive b
print("Tip promenljive b:", type(b))

# ponovna provera tipa niza a
print("Tip niza a:", type(a))

# ispis rezultata (opciono)
print("b =", b)
print("a =", a)

```

- Naći medijanu niza *a*

```python

a = [90, 80, 40, 70, 10]  # trenutni niz a

a_sortirano = sorted(a)
n = len(a_sortirano)

if n % 2 == 1:
    medijana = a_sortirano[n // 2]
else:
    medijana = (a_sortirano[n // 2 - 1] + a_sortirano[n // 2]) / 2

print("Medijana:", medijana)
```
