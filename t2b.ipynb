4. U sledećoj ćeliji su dati igrači koji igraju kolo. Izvršite je bez izmena, a onda napišite kod ispod koji spaja sve igrače, tako da se drže za ruke (primer: (0-0)\\/(o.o))

```python
igraci=["(-.-)","(^.^)","(*^*)","(o_o)","(0-0)"]
```

```python
# Ćelija koja se izvršava bez izmena
igraci = ["0-0", "o.o", "x-x"]

# Spajanje igrača tako da se "drže za ruke"
drze_za_ruke = r"\/".join([f"({i})" for i in igraci])

print(drze_za_ruke)
```

5. Ispitati promenljivu *procenat* iz drugog zadatka:
- ako je promenljiva *procenat* ispod 100, ispisati "Manje od dozvoljenog broja testova u jednoj nedelji",
- 100 ispisati "Ispunjen kapacitet testova"
- preko 100, ispisati  "Više od dozvoljenog broja testova u jednoj nedelji"

```python
# Primer promenljive procenat iz prethodnog zadatka
procenat = 150  # primer vrednosti

if procenat < 100:
    print("Manje od dozvoljenog broja testova u jednoj nedelji")
elif procenat == 100:
    print("Ispunjen kapacitet testova")
else:
    print("Više od dozvoljenog broja testova u jednoj nedelji")
```
