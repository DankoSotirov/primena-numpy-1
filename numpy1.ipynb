4a. Proveriti dimenzije te matrice, pomoću funkcije ndim, kao i shape i len.

```python

jedinicna = np.eye(3)

# Broj dimenzija
print("Broj dimenzija (ndim):", jedinicna.ndim)

# Dimenzije niza (shape)
print("Dimenzije niza (shape):", jedinicna.shape)

# Dužina niza (len) – vraća broj redova
print("Dužina niza (len):", len(jedinicna))
```

7. Iskoristiti niz iz prethodnog zadatka kao argument pravljenje niza *y*, koji sadrži sin(x), za svaki pojedinačni broj iz niza *x*.

```python
x = np.linspace(1, 7, 20)

# Napraviti niz y sa sin(x) za svaki element iz niza x
y = np.sin(x)

print(y)
```

8. Iskoristiti nizove *x* i *y*  za crtanje grafika funkcije sinus, pomoću paketa *matplotlib.pyplot* (funkcija *plt.plot(x,y)*)

```python
import numpy as np
import matplotlib.pyplot as plt

# Niz x i y
x = np.linspace(1, 7, 20)
y = np.sin(x)

# Crtanje grafika
plt.plot(x, y)
plt.title("Graf funkcije sin(x)")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.grid(True)
plt.show()
```
