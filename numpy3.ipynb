# Rešavanje linearnih jednačina pomoću numpy paketa. Iscrtavanje pomoću matplotlib paketa.

1. Napraviti matricu 3x3 M koja predstavlja koeficijente linearne jednačine, i vektor c koji predstavlja faktore linearne jednačine.


$x+2y-3z=-4 $

$-5y+7z=11$

$5y-4z=-2$


```python
import numpy as np

# Matrica koeficijenata
M = np.array([
    [1, 2, -3],   # x + 2y - 3z = -4
    [0, -5, 7],   # 0*x - 5y + 7z = 11
    [0, 5, -4]    # 0*x + 5y - 4z = -2
])

# Vektor faktora (desna strana jednačina)
c = np.array([-4, 11, -2])

print("Matrica M:\n", M)
print("Vektor c:\n", c)
```

2. Izračunati determinantu sistema. Da li je sistem rešiv?

```python
import numpy as np

# Matrica koeficijenata
M = np.array([
    [1, 2, -3],
    [0, -5, 7],
    [0, 5, -4]
])

# Determinanta
determinanta = np.linalg.det(M)
print("Determinanta:", determinanta)
```

3. Rešiti linearnu jednačinu i ispisati sva rešenja. Proveriti rešenja izvršavanjem ćelije ispod sledeće.

```python
import numpy as np

# Matrica koeficijenata
M = np.array([
    [1, 2, -3],
    [0, -5, 7],
    [0, 5, -4]
])

# Vektor desne strane
c = np.array([-4, 11, -2])

# Rešenje sistema
x = np.linalg.solve(M, c)
print("Rešenja sistema (x, y, z):", x)
```

4. Iscrtati funkcije na intervalu [0,1] koristeći pyplot:
 - log x
 - sin x
 - cos x
 - e^x
 - 10^x
 - x^2


```python
import numpy as np
import matplotlib.pyplot as plt

# Interval [0, 1] sa 100 tačaka
x = np.linspace(0.001, 1, 100)  # počinjemo od 0.001 da izbegnemo log(0)

# Definisanje funkcija
f_log = np.log(x)         # log(x)
f_sin = np.sin(x)         # sin(x)
f_cos = np.cos(x)         # cos(x)
f_exp = np.exp(x)         # e^x
f_10x = 10**x             # 10^x
f_x2 = x**2               # x^2

# Crtanje funkcija
plt.plot(x, f_log, label='log(x)')
plt.plot(x, f_sin, label='sin(x)')
plt.plot(x, f_cos, label='cos(x)')
plt.plot(x, f_exp, label='e^x')
plt.plot(x, f_10x, label='10^x')
plt.plot(x, f_x2, label='x^2')

# Naslov i oznake osa
plt.title("Različite funkcije na intervalu [0,1]")
plt.xlabel("x")
plt.ylabel("f(x)")

# Legenda i mreža
plt.legend()
plt.grid(True)

# Prikaz grafa
plt.show()
```

5. Iscrtati sinus i kosinus na intervalu [-5,5]. Obeležiti x i y ose i dodati legendu.

```python
import numpy as np
import matplotlib.pyplot as plt

# Interval [-5, 5] sa 200 tačaka
x = np.linspace(-5, 5, 200)

# Funkcije
y_sin = np.sin(x)
y_cos = np.cos(x)

# Crtanje funkcija
plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_cos, label='cos(x)')

# Naslov i oznake osa
plt.title("Sinus i Kosinus na intervalu [-5,5]")
plt.xlabel("x")
plt.ylabel("f(x)")

# Legenda i mreža
plt.legend()
plt.grid(True)

# Prikaz grafa
plt.show()
```
